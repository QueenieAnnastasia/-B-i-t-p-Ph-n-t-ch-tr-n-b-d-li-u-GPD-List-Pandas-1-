### 1. Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.


```python
import pandas as pd
df = pd.read_csv('GDPlist.csv', encoding = 'ISO-8859-1')
df
```

### 2. Việt hóa tên các cột trong bảng dữ liệu: Country -> Nuoc; Continent -> Chauluc; GDP (millions of US$) -> GDP (trieu $)


```python
df = df.rename(columns = {'Country':'Nuoc','Continent':'Chauluc','GDP (millions of US$)':'GDP (trieu$)'})
df
```

### 3. Chèn thêm một cột “Thanhpho” vào sau cột “Nuoc”, giá trị ban đầu là giá trị của cột “Nuoc”    


```python
df['Thanhpho']=df.loc[:,'Nuoc']
df
```

### 4. Trong cột Thanhpho, thay giá trị Vietnam thành Hanoi; Làm tương tự với các nước còn lại.


```python
df['Thanhpho'].replace('Vietnam','Hanoi', inplace = True)
df
```

```python
df.loc['Thanhpho':].replace('Vietnam','Hanoi', inplace = True)
df
```

### 5. Xóa các bản ghi có Chauluc là ‘Asia’


```python
df = df.loc[df['Chauluc'] != 'Asia']
df
```

### 6. Xoá các bản ghi có GDP < 300000

```python
df = df.loc[df['GDP (trieu$)'] >= 30000]
df
```
